# Ladda in filer från drive

Här är steg för steg hur man laddar in filer smidigt från drive

## 1. Installera och importera bibliotek

```
!pip3 install PyDrive
import os
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials  
```

## 2. Autentisera
Sedan måste man autentisera sig genom att köra följande koden nedan. Detta genererar en länk som tar en till en sida där man loggar in med sitt google-konto för att få en bekräftelsekod man måste klistra in och skriva in i ett fält som genereras av koden.

```
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

## 3. Ladda in filerna

Nedan är en exempelfil. För att göra detta med en ny fil ska man gå in på drive och högerklicka på det man vill ha därifrån (funkar bäst med zippad fil) och välja: hämta delbar länk. Då får man länk på följande format: `https://drive.google.com/open?id=FILE_ID`. Nedan ska man klistra in `FILE_ID` i `download = drive.CreateFile({'id': 'FILE_ID'})`. Sedan måste man skriva filnamnet på filen man vill hämta i `download.GetContentFile('FILE_NAME')`. Nedan är ett exempel på hur man kan göra, koden nedan laddar ner en satellitbild.

```
download = drive.CreateFile({'id': '1glwgYMyMr2P22euPjFtcIgCd8UggpQc-'})
download.GetContentFile('S2A_MSIL1C_20180731T102021_N0206_R065_T33UUB_20180731T123616.zip')
!ls
```

## 4. Packa upp zippade filer

Eftersom vi hämtade zippade filer måste de packas upp. Detta görs på olika sätt beroende på vilket format det är. Nedan följer exempel på hur man gör för `.zip` och `.tar.gz`.

```
#Packa upp .zip-fil
import zipfile
zip_ref = zipfile.ZipFile('S2A_MSIL1C_20180731T102021_N0206_R065_T33UUB_20180731T123616.zip')
zip_ref.extractall()
zip_ref.close()
```

```
#Packa upp .tar.gz-fil
import tarfile
tf = tarfile.open('S2A_MSIL1C_20180731T102021_N0206_R065_T33UUB_20180731T123616.tar.gz')
tf.extractall()
```

## 5. Ladda ned bearbetade filer
När man har kört sitt program och kanske genererat nya filer vill man kanske spara dessa. Nedan visas hur man antingen hämtar ner filer till sin dator och sparar dem lokalt, eller laddar upp dem på sin drive.

```
#Ladda ned filerna till datorn
from google.colab import files
files.download('example_file.ex') 
```

För att ladda upp dem på drive skriver du flöjande kod där `DRIVE.txt` är var du vill skriva filen till. `FILE_ON_COLAB.txt` är vad filen är sparad som i den nuvarande colab-sessionen.

```
#Ladda upp filerna på drive
upload = drive.CreateFile({'title': 'DRIVE.txt'})
upload.SetContentFile('FILE_ON_COLAB.txt')
upload.Upload()
```
